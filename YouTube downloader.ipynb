```python
#@title 入力画面
url = "https://youtu.be/iEjPvcFsFmI?si=NQlVhF95pXpgfxoN" # @param {type:"string"}
format = "best" # @param ["best", "1080p", "720p", "540p", "480p", "360p"]
```

```python

#@title エンコード
import yt_dlp
from yt_dlp import YoutubeDL  # 忘れている場合はインポートが必要
if format == "mp3":
        ydl_opts = {
            'format': 'bestaudio/best',
            'postprocessors': [
                {'key': 'FFmpegExtractAudio',
                 'preferredcodec': 'mp3',
                 'preferredquality': '192'},
                {'key': 'FFmpegMetadata'},
            ],
        }
elif format == "original":
        ydl_opts = {
            'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]',
            'outtmpl': 'download.mp4',  # ファイル名のテンプレート
        }
else:
        ydl_opts = {
            'format': f'bestvideo[height={format}][ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]',
            'outtmpl': 'download.mp4',  # ファイル名のテンプレート
        }

    # ダウンロード処理
with YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])
```
